```python
<VSCode.Cell language="markdown">
# Build a GPT-style tokenizer from scratch

This notebook walks through a complete educational tokenizer implementation in pure Python.
We build incrementally from a simple regex tokenizer to a mini GPT-style tokenizer with vocabulary, special tokens, byte pair encoding, visualization, and comparisons with real tokenizers.

Each section contains explanations, runnable code, and example outputs.
</VSCode.Cell>
<VSCode.Cell language="markdown">
## Part 1 — Basic Regex Tokenizer

We start by loading raw text from a file and splitting it into meaningful pieces using regular expressions.
This stage is what many early NLP systems did before subword tokenizers were invented.
</VSCode.Cell>
<VSCode.Cell language="python">
from pathlib import Path

# Create a sample raw text file for this tutorial.
# In a real workflow, this would be an existing dataset file.
raw_text_path = Path("sample_text.txt")
raw_text_path.write_text(
    "Hello, world. Is this-- a test?\n"
    "Tokenization is the first step in text processing.\n"
    "It splits words, punctuation, and special symbols.\n"
)

# Load the raw text file into Python.
raw_text = raw_text_path.read_text()
print("--- Raw Text ---")
print(raw_text)
</VSCode.Cell>
<VSCode.Cell language="markdown">
### Why tokenization matters

- Raw text is a sequence of characters, but models operate on discrete units called tokens.
- Tokenization converts text into those units.
- Punctuation matters because `Hello,` and `Hello` are different strings with different meanings for parsers and language models.
- Whitespace must be handled carefully to avoid creating meaningless tokens like `" "`, and to preserve the text structure.
</VSCode.Cell>
<VSCode.Cell language="python">
import re

# Define a simple regex-based tokenizer.
# The regex splits on words, punctuation, double dashes, and special symbols.
# The order of alternatives is important: longer tokens like '--' must come before single characters.
TOKEN_PATTERN = re.compile(r"--|\w+|[.,!?;:\"'()\[\]{}]|[^\s\w]")


def regex_tokenize(text: str) -> list[str]:
    """Tokenize text into words, punctuation, and symbols, preserving case."""
    tokens = TOKEN_PATTERN.findall(text)
    return tokens

# Tokenize the raw text and show example output.
example_tokens = regex_tokenize(raw_text)
print("Example tokens:")
print(example_tokens)

# Another explicit example from the problem description.
example_sentence = "Hello, world. Is this-- a test?"
print("\nInput:", example_sentence)
print("Output:", regex_tokenize(example_sentence))
</VSCode.Cell>
<VSCode.Cell language="markdown">
### What this tokenizer does

- `\w+` matches words and numbers.
- `--` matches double dashes as one token.
- `[.,!?;:\"'()\[\]{}]` matches common punctuation.
- `[^\s\w]` catches any remaining non-space, non-word symbols.

This keeps capitalization intact and removes redundant whitespace by not producing space tokens.
</VSCode.Cell>
<VSCode.Cell language="markdown">
## Part 2 — Vocabulary Creation

A vocabulary maps each unique token to a numeric ID. Models cannot work directly on text, so token IDs are the bridge between strings and numbers.
</VSCode.Cell>
<VSCode.Cell language="python">
from typing import Tuple


def build_vocab(tokens: list[str]) -> Tuple[dict[str, int], dict[int, str]]:
    """Build a vocabulary mapping from tokens to IDs and back."""
    unique_tokens = sorted(set(tokens))
    token_to_id = {token: idx for idx, token in enumerate(unique_tokens)}
    id_to_token = {idx: token for token, idx in token_to_id.items()}
    return token_to_id, id_to_token

# Build the vocabulary from the example tokens.
token_to_id, id_to_token = build_vocab(example_tokens)

print("Vocabulary size:", len(token_to_id))
print("Sample token -> id entries:")
for token in list(token_to_id)[:15]:
    print(f"  {token!r}: {token_to_id[token]}")
</VSCode.Cell>
<VSCode.Cell language="markdown">
### Why token IDs exist

- Transformers and neural networks operate on numeric arrays, not raw text.
- Token IDs are integer representations of discrete text units.
- The mapping itself has no semantic meaning; the model learns meaning from token sequences during training.
- The actual numbers are arbitrary, but consistency is essential.
</VSCode.Cell>
<VSCode.Cell language="markdown">
## Part 3 — Simple Tokenizer Class

We now wrap the tokenization and vocabulary lookup logic into a reusable tokenizer class.
This class can encode text to IDs and decode IDs back to human-readable text.
</VSCode.Cell>
<VSCode.Cell language="python">
import re
from dataclasses import dataclass


def clean_decoded_text(text: str) -> str:
    """Fix spacing around punctuation after decoding."""
    text = re.sub(r"\s+([.,!?;:\)\]\}])", r"\1", text)
    text = re.sub(r"([\(\[\{])\s+", r"\1", text)
    text = re.sub(r"\s+--\s+", "--", text)
    return text.strip()


@dataclass
class SimpleTokenizerV1:
    token_to_id: dict[str, int]
    id_to_token: dict[int, str]

    def encode(self, text: str) -> list[int]:
        tokens = regex_tokenize(text)
        return [self.token_to_id[token] for token in tokens]

    def decode(self, ids: list[int]) -> str:
        tokens = [self.id_to_token[token_id] for token_id in ids]
        decoded = " ".join(tokens)
        return clean_decoded_text(decoded)


# Instantiate and test the simple tokenizer.
base_tokenizer = SimpleTokenizerV1(token_to_id=token_to_id, id_to_token=id_to_token)
encoded = base_tokenizer.encode(example_sentence)
decoded = base_tokenizer.decode(encoded)

print("Encoded IDs:", encoded)
print("Decoded text:", decoded)
print("Round-trip preserved?:", decoded == example_sentence)
</VSCode.Cell>
<VSCode.Cell language="markdown">
### Round-trip tokenization

The tokenizer should convert text to IDs and back without changing the meaning.
Our simple decoder re-joins tokens and removes extra spaces around punctuation so the output is readable.
</VSCode.Cell>
<VSCode.Cell language="markdown">
## Part 4 — Unknown Token Handling

Real tokenizers must handle words they have never seen before.
We add special tokens for unknown text and end-of-text markers.
</VSCode.Cell>
<VSCode.Cell language="python">
UNKNOWN_TOKEN = "<|unk|>"
END_OF_TEXT_TOKEN = "<|endoftext|>"


def build_vocab_with_special(tokens: list[str]) -> Tuple[dict[str, int], dict[int, str]]:
    sorted_tokens = sorted(set(tokens))
    sorted_tokens = [UNKNOWN_TOKEN, END_OF_TEXT_TOKEN] + sorted_tokens
    token_to_id = {token: idx for idx, token in enumerate(sorted_tokens)}
    id_to_token = {idx: token for token, idx in token_to_id.items()}
    return token_to_id, id_to_token


class SimpleTokenizerV2(SimpleTokenizerV1):
    unk_token: str = UNKNOWN_TOKEN

    def encode(self, text: str) -> list[int]:
        tokens = regex_tokenize(text)
        return [self.token_to_id.get(token, self.token_to_id[self.unk_token]) for token in tokens]


vocab2_token_to_id, vocab2_id_to_token = build_vocab_with_special(example_tokens)
unknown_tokenizer = SimpleTokenizerV2(token_to_id=vocab2_token_to_id, id_to_token=vocab2_id_to_token)

# Test with an unknown token.
unknown_sentence = "This example contains unseenword and punctuation!"
encoded_unknown = unknown_tokenizer.encode(unknown_sentence)
print("Encoded IDs:", encoded_unknown)
print("Decoded text:", unknown_tokenizer.decode(encoded_unknown))
print("Unknown token ID:", vocab2_token_to_id[UNKNOWN_TOKEN])
</VSCode.Cell>
<VSCode.Cell language="markdown">
### Why unknown tokens are necessary

- OOV (out-of-vocabulary) words occur when the tokenizer encounters a word not in its vocabulary.
- Without a fallback, the tokenizer would crash or fail to represent text.
- Special tokens let models handle arbitrary input and control structure.
- GPT-style models use tokens like `<|endoftext|>` to mark the end of a text sequence.
</VSCode.Cell>
<VSCode.Cell language="markdown">
## Part 5 — Byte Pair Encoding (BPE)

Word-level vocabularies struggle with rare or unknown words. BPE learns subword units by merging frequent character pairs.
This is a simplified version of the algorithm used in many GPT-style tokenizers.
</VSCode.Cell>
<VSCode.Cell language="python">
from collections import Counter


def get_vocab_from_corpus(corpus: list[str]) -> Counter[str]:
    """Create an initial vocabulary from words represented as character sequences."""
    vocab = Counter()
    for sentence in corpus:
        for word in regex_tokenize(sentence):
            vocab[" ".join(list(word)) + " </w>"] += 1
    return vocab


def get_stats(vocab: Counter[str]) -> Counter[tuple[str, str]]:
    """Count frequencies of symbol pairs in the vocabulary."""
    pairs = Counter()
    for word, freq in vocab.items():
        symbols = word.split()
        for i in range(len(symbols) - 1):
            pairs[(symbols[i], symbols[i + 1])] += freq
    return pairs


def merge_vocab(pair: tuple[str, str], vocab_in: Counter[str]) -> Counter[str]:
    """Merge the most frequent pair in the vocabulary."""
    bigram = re.escape(" ".join(pair))
    pattern = re.compile(r"(?<!\S)" + bigram + r"(?!\S)")
    vocab_out = Counter()
    replacement = "".join(pair)
    for word, freq in vocab_in.items():
        new_word = pattern.sub(replacement, word)
        vocab_out[new_word] = freq
    return vocab_out


def train_bpe(corpus: list[str], num_merges: int) -> tuple[Counter[str], list[tuple[str, str]]]:
    """Train a simplified BPE model and return the final vocab and merge rules."""
    vocab = get_vocab_from_corpus(corpus)
    merges = []
    for i in range(num_merges):
        pairs = get_stats(vocab)
        if not pairs:
            break
        most_common = pairs.most_common(1)[0][0]
        merges.append(most_common)
        vocab = merge_vocab(most_common, vocab)
        print(f"Merge {i + 1}: {most_common}")
    return vocab, merges


# Train BPE on a simple corpus.
small_corpus = [
    "hello world",
    "hello there",
    "hell of a world",
    "help me with hello",
    "where is the yellow light"
]
final_vocab, merge_rules = train_bpe(small_corpus, num_merges=10)

print("\nFinal vocabulary entries:")
for item, freq in final_vocab.most_common(15):
    print(f"  {item}: {freq}")

print("\nLearned merges:", merge_rules)
</VSCode.Cell>
<VSCode.Cell language="python">

def apply_bpe(word: list[str], merges: list[tuple[str, str]]) -> list[str]:
    """Encode a single word using learned BPE merge rules."""
    symbols = word[:]
    while True:
        pairs = [(symbols[i], symbols[i + 1]) for i in range(len(symbols) - 1)]
        merge_candidates = {pair: idx for idx, pair in enumerate(merges) if pair in pairs}
        if not merge_candidates:
            break
        pair_to_merge = min(merge_candidates, key=merge_candidates.get)
        new_symbols = []
        i = 0
        while i < len(symbols):
            if i < len(symbols) - 1 and (symbols[i], symbols[i + 1]) == pair_to_merge:
                new_symbols.append(symbols[i] + symbols[i + 1])
                i += 2
            else:
                new_symbols.append(symbols[i])
                i += 1
        symbols = new_symbols
    return symbols


def encode_bpe(text: str, merges: list[tuple[str, str]]) -> list[str]:
    tokens = regex_tokenize(text)
    encoded = []
    for token in tokens:
        token_symbols = list(token) + ["</w>"]
        merged = apply_bpe(token_symbols, merges)
        encoded.extend(merged)
    return encoded

# Example BPE encoding.
example_bpe_sentence = "hello yellow world"
print("BPE tokens:", encode_bpe(example_bpe_sentence, merge_rules))
</VSCode.Cell>
<VSCode.Cell language="markdown">
### Why BPE is powerful

- Word-level vocabularies cannot cover every possible word.
- BPE builds frequently occurring subword units from characters.
- This allows the tokenizer to represent rare words by combining subword pieces.
- GPT tokenizers use variants of BPE to efficiently encode text with a manageable vocabulary size.
</VSCode.Cell>
<VSCode.Cell language="markdown">
## Part 6 — Mini GPT-style Tokenizer

Now we combine everything into a more practical tokenizer class with options for lowercasing, special tokens, regex configuration, JSON save/load, padding, and truncation.
</VSCode.Cell>
<VSCode.Cell language="python">
import json


@dataclass
class MiniGPTTokenizer:
    token_to_id: dict[str, int]
    id_to_token: dict[int, str]
    unk_token: str = UNKNOWN_TOKEN
    pad_token: str = " "
    eos_token: str = END_OF_TEXT_TOKEN
    lowercase: bool = False
    pattern: re.Pattern = TOKEN_PATTERN

    def __post_init__(self):
        if self.pad_token not in self.token_to_id:
            pad_id = len(self.token_to_id)
            self.token_to_id[self.pad_token] = pad_id
            self.id_to_token[pad_id] = self.pad_token

    def tokenize(self, text: str) -> list[str]:
        if self.lowercase:
            text = text.lower()
        return self.pattern.findall(text)

    def encode(self, text: str, add_eos: bool = False) -> list[int]:
        tokens = self.tokenize(text)
        ids = [self.token_to_id.get(token, self.token_to_id[self.unk_token]) for token in tokens]
        if add_eos:
            ids.append(self.token_to_id[self.eos_token])
        return ids

    def decode(self, ids: list[int]) -> str:
        tokens = [self.id_to_token.get(token_id, self.unk_token) for token_id in ids]
        return clean_decoded_text(" ".join(tokens))

    def batch_encode(self, texts: list[str], add_eos: bool = False) -> list[list[int]]:
        return [self.encode(text, add_eos=add_eos) for text in texts]

    def batch_decode(self, id_lists: list[list[int]]) -> list[str]:
        return [self.decode(ids) for ids in id_lists]

    def pad(self, ids: list[int], length: int) -> list[int]:
        pad_id = self.token_to_id[self.pad_token]
        return ids[:length] + [pad_id] * max(0, length - len(ids))

    def truncate(self, ids: list[int], max_length: int) -> list[int]:
        return ids[:max_length]

    def save(self, path: Path) -> None:
        with path.open("w", encoding="utf-8") as f:
            json.dump({"token_to_id": self.token_to_id}, f, ensure_ascii=False, indent=2)

    @classmethod
    def load(cls, path: Path, lowercase: bool = False, pattern: re.Pattern = TOKEN_PATTERN) -> "MiniGPTTokenizer":
        with path.open("r", encoding="utf-8") as f:
            data = json.load(f)
        token_to_id = data["token_to_id"]
        id_to_token = {int(idx): token for token, idx in token_to_id.items()}
        return cls(token_to_id=token_to_id, id_to_token=id_to_token, lowercase=lowercase, pattern=pattern)


# Build a tokenizer with special tokens and a padding token.
mini_tokenizer = MiniGPTTokenizer(token_to_id=vocab2_token_to_id.copy(), id_to_token=vocab2_id_to_token.copy())
print("PAD token ID:", mini_tokenizer.token_to_id[mini_tokenizer.pad_token])

text = "Hello, new tokenizer!"
ids = mini_tokenizer.encode(text, add_eos=True)
print("Encoded:", ids)
print("Decoded:", mini_tokenizer.decode(ids))

batch_ids = mini_tokenizer.batch_encode([text, example_sentence], add_eos=True)
print("Batch encoded:", batch_ids)
print("Padded:", mini_tokenizer.pad(batch_ids[0], length=12))
</VSCode.Cell>
<VSCode.Cell language="markdown">
### Practical tokenizer features

- `lowercase` lets you normalize input before tokenization.
- `special tokens` support unknown and padding values.
- `batch_encode`/`batch_decode` handle multiple texts efficiently.
- `save` and `load` let you persist the vocabulary as JSON.
- `padding` and `truncation` are important for batching input in model training and inference.
</VSCode.Cell>
<VSCode.Cell language="markdown">
## Part 7 — Visualization and Analysis

We can analyze the vocabulary and token frequencies visually.
This helps understand which tokens are common and which are rare.
</VSCode.Cell>
<VSCode.Cell language="python">
import matplotlib.pyplot as plt
from collections import Counter

# Compute token frequencies from our example text.
token_counts = Counter(regex_tokenize(raw_text))
most_common = token_counts.most_common(15)
print("Most common tokens:")
for token, count in most_common:
    print(f"  {token!r}: {count}")

# Plot the most common tokens.
plt.figure(figsize=(10, 5))
plt.bar([token for token, _ in most_common], [count for _, count in most_common])
plt.title("Top 15 Token Frequencies")
plt.xlabel("Token")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.show()

# Vocabulary statistics.
all_tokens = sorted(token_counts)
longest_tokens = sorted(all_tokens, key=len, reverse=True)[:10]
rare_tokens = [token for token, count in token_counts.items() if count == 1]

print("\nLongest tokens:", longest_tokens)
print("Number of rare tokens (count == 1):", len(rare_tokens))
print("Rare tokens sample:", rare_tokens[:10])
</VSCode.Cell>
<VSCode.Cell language="markdown">
### What visual analysis tells us

- Frequency histograms show which tokens the model will see most often.
- Long tokens reveal punctuation or multi-character words.
- Rare tokens highlight where a subword tokenizer like BPE can help by breaking words into more common pieces.
</VSCode.Cell>
<VSCode.Cell language="markdown">
## Part 8 — Performance Improvements

Pure Python tokenizers are easy to understand, but they can be slow for large datasets. Production tokenizers are often written in Rust or C++ for speed.
</VSCode.Cell>
<VSCode.Cell language="python">
from functools import lru_cache

# Example optimization: cache the results of tokenization for repeated inputs.
@lru_cache(maxsize=256)
def cached_tokenize(text: str) -> tuple[str, ...]:
    return tuple(regex_tokenize(text))

print("Cached tokenization example:", cached_tokenize("Hello, world."))

# Explain complexity:
print("\nComplexity notes:")
print("- Tokenization is typically O(n) in the length of the input text.")
print("- Vocabulary lookup is O(1) per token using a dictionary.")
print("- BPE training is more expensive because it recomputes pair frequencies.")
print("- Production tokenizers use compiled code to reduce Python overhead and memory allocations.")
</VSCode.Cell>
<VSCode.Cell language="markdown">
### Why performance matters

- Tokenizers are often the first stage in a pipeline, so their cost affects overall latency.
- Lookup in dictionaries is fast, but Python function overhead and regex work can dominate.
- Compiled implementations reduce memory churn and improve throughput.
</VSCode.Cell>
<VSCode.Cell language="markdown">
## Part 9 — Compare with Real Tokenizers

We compare our tokenizer with GPT-2, TikToken, and SentencePiece if those libraries are available.
The code below attempts to import these libraries and prints the tokenization results for the same example sentence.
</VSCode.Cell>
<VSCode.Cell language="python">
example_compare = "Hello, world. This is a comparison between tokenizers."
print("Example sentence:", example_compare)

libraries = ["transformers", "tiktoken", "sentencepiece"]
for lib in libraries:
    try:
        if lib == "transformers":
            from transformers import GPT2TokenizerFast
            tokenizer = GPT2TokenizerFast.from_pretrained("gpt2")
            tokens = tokenizer.tokenize(example_compare)
            ids = tokenizer.encode(example_compare)
            print("\nGPT-2 tokenizer:")
            print(" Tokens:", tokens)
            print(" IDs:", ids)
            print(" Count:", len(tokens))
        elif lib == "tiktoken":
            import tiktoken
            enc = tiktoken.get_encoding("gpt2")
            ids = enc.encode(example_compare)
            tokens = [enc.decode([token_id]) for token_id in ids]
            print("\nTikToken tokenizer:")
            print(" Tokens:", tokens)
            print(" IDs:", ids)
            print(" Count:", len(ids))
        elif lib == "sentencepiece":
            import sentencepiece as spm
            print("\nSentencePiece: library installed. Please load a model to compare tokenization.")
    except Exception as e:
        print(f"\n{lib} not available or failed to load: {e}")
</VSCode.Cell>
<VSCode.Cell language="markdown">
### Real tokenizer tradeoffs

- GPT-2 tokenization uses a BPE vocabulary learned on large corpora.
- TikToken is designed for efficient OpenAI-style tokenization and usually matches GPT-2 exactly.
- SentencePiece is another subword tokenizer that can operate in either unigram or BPE mode.
- Our custom tokenizer is for education and does not match production-level text coverage or efficiency.
</VSCode.Cell>
<VSCode.Cell language="markdown">
## Exercises and Extensions

1. Add support for a full Byte Pair Encoding tokenizer class with vocabulary saving.
2. Extend the tokenizer to support a `bos_token` and `mask_token`.
3. Build a tokenizer evaluation function that compares token counts and split quality across multiple sentences.
4. Add a custom rule so URLs and email addresses are tokenized as single units.
5. Use a larger text corpus and measure how vocabulary size changes when adding BPE merges.

This notebook is a foundation for understanding how GPT-style tokenizers are built from the ground up.
</VSCode.Cell>
```
